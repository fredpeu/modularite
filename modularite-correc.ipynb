## PREMIÈRE PARTE: MODULES

Nous connaissons bien les fonctions `print` et `input`. Ce sont deux fonctions prédéfinies dans le langage Python, On appelle ce type de fonction des **fonctions natives** au langage Python , elles sont directement utilisables.

D'autres fonctions ont été développées et testées par différents programmeurs mais ne sont pas directement utilisables. Ces fonctions sont regroupées dans des modules, appelés aussi bibliothèques. Nous en avons déjà reontré un certain nombres: `folio`, `math`, `random`, `turtle`, `csv` ou encore `pygame`. 

### Import d´un module en Phyton et docstring

Prenons l´exemple du module `random` et utilisons la fonction (native) `help` pour avoir toutes les informations sur le module:

```python
import random as rdm
# affichage de la documentation
help(rdm)
```

La fonction donne accès à la documentation complète du module: la **docstring**.

> Affichez la **docstring** de la fonction `randint`:

```python
# affichage
help(rdm.randint)
```

Pour créer la docstring d´une **fonction**, on insère les informations en début de bloc d´instructions entre une paire de trois guillemets.

> Créez la documentation de cette fonction et affichez sa docstring:

```python
def double(x:int)->int:
    '''
    - Je prends un entier en entrée.
    - Je retourne un entier: le double de la valeur passée en paramètre.
    '''
    return 2*x

help(double)
```

Pour créer la docstring d´un **module**, on insère les informations en début de fichier entre une paire de trois guillemets.
Dans le jeu `pygame`, nous avions créé un module `utils`que vous pouvez maintenant enrichir de sa **docstring**.

## Les API

Une API (Application Programming Interface) est une interface de programmation d'application. Cette interface est destinée à être utilisée par des programmes. Elle est composée de fonctions, constantes, classes. 

### API web

Dans une API web, il y a un échange de données avec des formats qui se sont standardisés dans le temps. Il existe essentiellement deux formats **XML(Extensible Markup Language)** et **JSON (JavaSript Objet Notation)**.

En exemple, l´**IPMA** propose une API web qui récolte les informations météo stockées sur leur serveur. L´utilisateur de l´API peut l´utiliser pour créer une application qui affiche les prévisions météo au Portugal sur smartphone.

### Etude d´un exemple:

Nous allons tester l´utilisation de l´API qui rencense toutes les adresses connues en France.

Allons consulter la documentation de l´API à cette adresse : https://geo.api.gouv.fr/adresse

Pour accéder aux données de l´API, il faut faire des requêtes `https`. En Python, le module `requests`a été conçu pour cela:

```python
# import du module
import requests
```

Testons une première requête (`GET`):

```python
url1="https://api-adresse.data.gouv.fr/search/?q=2+rue+sousa+mendes" # &limit=10 si on veut limiter le nombre de résultats
rep1=requests.get(url1)
rep1
```

`rep1` est l´instance d´un objet contenant les informations.
Pour les visualiser:

```python
donnees=rep1.json()
donnees
```

Les informations sont stockées dans le format JSON qui apparait sous la forme d´un dictionnaire assez complexe.

### Tri des informations:


> Recherchez dans le dictionnaire où sont stockées les informations retournées et affichez-les. Sous quelle forme sont-elles stockées?

```python
donnees['features']
```

> Affichez le nombre de résultats retournés:

```python
len(donnees['features'])
```

> Affichez le premier résultat retourné:

```python
donnees['features'][0]
```

> Affichez le nom de la  ville concernée:

```python
donnees['features'][0]['properties']['city']
```

> Affichez le nom exact de la rue:

```python
donnees['features'][0]['properties']['name']
```

> Affichez les coordonnées GPS de cette adresse:

```python
donnees['features'][0]['geometry']['coordinates']
```

> Affichez le nom de toutes les villes qui on une adresse **2 rue Sousa Mendes**:

```python
 for i in range(len(donnees['features'])):
        print (donnees['features'][i]['properties']['city'])
```

Testons une dernière requête:

```python
url2="https://api-adresse.data.gouv.fr/reverse/?lon=3.2109&lat=49.6175&type=street"
rep2=requests.get(url2)
rep2
```

> Affichez le nom de la ville et de la rue où se trouve cette adresse...

```python
donnees2=rep2.json()
print(donnees2['features'][0]['properties']['name'],donnees2['features'][0]['properties']['city'])
```
